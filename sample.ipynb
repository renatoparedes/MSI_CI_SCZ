```python
t = 100
soa = 50
A, V, M = ci.run_dfi(t, soa)
```

```python
x = np.arange(0,t,0.01)
y = np.arange(180)
X,Y = np.meshgrid(x,y)

#fig,axs = plt.subplots(2, 2, figsize=(18, 18))

fig = plt.figure(figsize=(18,18))
ax1 = fig.add_subplot(232, projection='3d')
ax2 = fig.add_subplot(234, projection='3d')
ax3 = fig.add_subplot(236, projection='3d')


# Plot a 3D surface
ax3.plot_surface(X, Y, A)
ax3.set_title("Auditory", size=14)
ax2.plot_surface(X, Y, V)
ax2.set_title("Visual", size=14)
ax1.plot_surface(X, Y, M)
ax1.set_title("Multisensory", size=14)

fig.subplots_adjust(hspace=-0.8)
fig.subplots_adjust(wspace=0.2)
```
