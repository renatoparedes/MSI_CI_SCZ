```python
import ci_model as ci
import matplotlib.pyplot as plt
import numpy as np
```

```python
Y_a, Y_v, Y_m = ci.run_cuppini(100, pos_a=90,pos_v=90, Ea = 18.75, Ev = 0, noise=False) 
ci.plot_areas(Y_a, Y_v, Y_m)
```

```python
# Model run
t = 150
soa = 14
n = 90

A, V, M = ci.run_dfi(t, soa, N=n, pos_a=45,pos_v=45)

# Plotting

x = np.arange(0,t,0.01)
y = np.arange(n)
X,Y = np.meshgrid(x,y)

#fig,axs = plt.subplots(2, 2, figsize=(18, 18))

fig = plt.figure(figsize=(18,18))
ax1 = fig.add_subplot(232, projection='3d')
ax2 = fig.add_subplot(234, projection='3d')
ax3 = fig.add_subplot(236, projection='3d')


# Plot a 3D surface
ax3.plot_surface(X, Y, A)
ax3.set_title("Auditory", size=14)
ax2.plot_surface(X, Y, V)
ax2.set_title("Visual", size=14)
ax1.plot_surface(X, Y, M)
ax1.set_title("Multisensory", size=14)

ax3.set_zlabel("Neural activity")
ax3.set_ylabel("Position (Degrees)")
ax3.set_xlabel("Time (ms)")

ax1.set_zlabel("Neural activity")
ax1.set_ylabel("Position (Degrees)")
ax1.set_xlabel("Time (ms)")

ax2.set_zlabel("Neural activity")
ax2.set_ylabel("Position (Degrees)")
ax2.set_xlabel("Time (ms)")

fig.subplots_adjust(hspace=-0.8)
fig.subplots_adjust(wspace=0.2)
```
